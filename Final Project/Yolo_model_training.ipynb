```python
import os
import cv2
import matplotlib.pyplot as plt

# Assuming you have the 'model' object already defined

folder_path = '/content/drive/MyDrive/wild_life/datasets/test/images'   # Set path accordingly
image_files = os.listdir(folder_path)

# Create subplots with 2 rows and 2 columns
num_images = len(image_files)
num_rows = 2
num_cols = 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 8))
index = 1
for i in range(num_rows):
    for j in range(num_cols):
        index = index + index + 2
        #  (i * num_cols + j)
        if index < num_images:
            image_file = image_files[index]
            image_path = os.path.join(folder_path, image_file)
            img = cv2.imread(image_path)


            results = model(img)

            res_plotted = results[0].plot()
            output_img = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)

            axes[i, j].imshow(output_img)
            axes[i, j].axis('off')  # Turn off axis labels

plt.show()
```
